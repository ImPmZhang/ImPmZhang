```python
from faker import Faker
import pandas as pd
import numpy as np
import random

fa = Faker(locale='zh_CN')

num = 10

# 初始化一个空列表来存储随机时间
random_times = []

# 循环生成随机时间
for _ in range(num):
    # 生成小时（0-23）
    hour = random.randint(0, 23)
    # 生成分钟（0-59）
    minute = random.randint(0, 59)
    # 将小时和分钟转换为字符串并组合成时间格式
    time_str = f"{hour:02d}:{minute:02d}"
    # 将随机时间添加到列表中
    random_times.append(time_str)
    

pd.DataFrame({
    "商户名称":[fa.company() for _ in range(num) ],
    "操作人姓名":[fa.name() for _ in range(num)],
    "商品编号":[fa.pystr(max_chars=10) for _ in range(num)],
    # "商品信息完善度":np.random.rand(1,num).tolist()
    "操作时间":random_times,
    "操作状态":[fa.random_element(['保存失败','保存成功','上架失败','上架成功']) for _ in range(num)]
})

```

```python
import datetime

def getrandometime():
    # 获取当前时间，并设置时间为当天早上8点
    now = datetime.datetime.now()
    today_morning_8 = now.replace(hour=8, minute=0, second=0, microsecond=0)
    
    
    # 随机选择这三天内的某一天早上8点
    random_morning_8 = today_morning_8 - datetime.timedelta(days=random.randint(0, 2))
    
    # 在该天的8点到20点之间随机选择一个时间
    hours = random.randint(8, 19)  # 8点到19点，因为20点不包括
    minutes = random.randint(0, 59)
    seconds = random.randint(0, 59)
    
    # 构造随机时间
    random_time = random_morning_8.replace(hour=hours, minute=minutes, second=seconds)
    
    # 格式化输出时间
    formatted_time = random_time.strftime('%Y-%m-%d %H:%M:%S')
    
    # print(formatted_time)
    return formatted_time

getrandometime()
```

```python

shanghu_num = 3
yuangong_num = 10
shangpin_num = 30
dongzuo_num = 150

shanghu = pd.DataFrame({
    "商户名称":[fa.company() for _ in range(shanghu_num) ],
    "商户ID":[x+1 for x in range(shanghu_num)]

})

yuangong = pd.DataFrame({
    "员工姓名":[fa.name() for _ in range(yuangong_num)],
    "商户ID":(random.randint(1,shanghu_num) for _ in range(yuangong_num)),
    "员工ID":[x+1 for x in range(yuangong_num)]
})

shangpin = pd.DataFrame({
    "商品编号":[fa.pystr(max_chars=10) for _ in range(shangpin_num)],
    "商品信息完善度":(random.random() for _ in range(shangpin_num)),
    "员工ID":(random.randint(1,yuangong_num) for _ in range(shangpin_num)),
    "商品ID":[x+1 for x in range(shangpin_num)]
})

dongzuo = pd.DataFrame({
    "操作时间":[getrandometime() for _ in range(dongzuo_num)],
    "操作状态":[fa.random_element(['保存失败','保存成功','上架失败','上架成功']) for _ in range(dongzuo_num)],
    "商品ID":(random.randint(1,shangpin_num) for _ in range(dongzuo_num)),
    "动作ID":[x+1 for x in range(dongzuo_num)]
})
df= (
pd.merge(dongzuo,shangpin,how='left', on='商品ID')
.merge(yuangong,how='left',on='员工ID')
.merge(shanghu,how='left',on='商户ID')
.drop(['商品ID','员工ID','商户ID'],axis=1)
.sort_values(by='操作时间')
.reset_index(drop=True)
)
print(df)
with pd.ExcelWriter("/Users/henry/Downloads/pandas/商品系统-输出.xlsx") as writer:
     shanghu.to_excel(writer,index=False,sheet_name='商户表',engine='openpyxl')
     yuangong.to_excel(writer,index=False,sheet_name='员工表',engine='openpyxl')
     shangpin.to_excel(writer,index=False,sheet_name='商品表',engine='openpyxl')
     dongzuo.to_excel(writer,index=False,sheet_name='动作表',engine='openpyxl')
     df.to_excel(writer,index=False,sheet_name='汇总表',engine='openpyxl')
          
```
