# 1. 信息处理
本数据完全是伪数据，需要进行数据处理

```python
import pandas as pd
import numpy as np
ex = pd.read_excel('/Users/henry/Downloads/pandas/商品系统-输出.xlsx',sheet_name='汇总表',dtype={
    "操作时间":'datetime64[ns]',
    '操作状态':'category',
    '商品编号':'category',
    '员工姓名':'category',
    '商户名称':'category'
})
ex
```

#  2. 事件定义
- 列名：事件类型
- 列值，上架已成功/上架待成功
## 处理思路，
- 时间窗口内，如果商品上架成功，则标记为“上架已成功”且抛掉首次成功之后的数据
- 如果上架未成功，则标记为“上架待成功”

```python
s = (
ex.assign( 事件类型 = ex.操作状态)
.sort_values('操作时间')
.groupby(['商品编号'],observed=True)
.apply(lambda gro: gro[:(gro.操作状态 == '上架成功').idxmax()+1] if (gro.操作状态 == '上架成功').any() else gro )
.reset_index(drop=True)

)
ss = (s.groupby(['商品编号'],observed=True)
.事件类型
.transform(lambda sj: "上架已成功"  if (sj=='上架成功').any() else '上架待成功')
)
s.事件类型 = ss
s
```

# 3.事件编号
这里直接使用 商品编号作为序号


```python
s['事件编号']=s['商品编号']
s
```

# 4. 计算时间差

```python

s= s.sort_values(['事件编号','操作时间'])
s['时间差']= s.groupby('事件编号').操作时间.diff().dt.total_seconds().div(3600, fill_value=0).round(1)
```

# 透视表

```python
pp = s.pivot_table(values='时间差',index=['事件类型','事件编号'],aggfunc=['sum','count','mean'])
pp = pp.reset_index().dropna()
pp.columns=['事件类型','事件编号','时间','次数','平均值']
print(pp)
```
